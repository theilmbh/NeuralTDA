```python
import neuraltda.simpComp as sc
import numpy as np
import scipy as sp
```

```python
# generate random max simps:
def gen_sc(n_ms, n_units, maxdim):
    maxsimps = []
    for idn in range(n_ms):
        ms_len = np.random.randint(maxdim)
        ms = tuple(np.random.choice(range(n_units), size=ms_len))
        maxsimps.append(ms)
    return maxsimps
        
```

```python
nmaxsimps = 1000
n_units = 32
maxdim = 5
ms1 = gen_sc(nmaxsimps, n_units, maxdim)
ms2 = gen_sc(nmaxsimps, n_units, maxdim)
```

```python
scg1 = sc.simplicialChainGroups(ms1)
scg2 = sc.simplicialChainGroups(ms2)
```

```python
%%time 
L1 = sc.sparse_laplacian(scg1, 1)
L2 = sc.sparse_laplacian(scg2, 1)
```

```python
%%time
kl = sc.sparse_KL_divergence(L1, L2, 1.0)
print(kl)
```

```python
%%time
JS = sc.sparse_JS_divergence(L1, L2, 1.0)
print(JS)
```
